Habiendo investigado ya la base de datos de "companies", decido montar mi sede en New York cerca de una empresa que lleva activa desde 2005 y haya generado mas de $1M

```python
companies_new_york = collection.find({"offices.city": "New York", "founded_year": {"$gt": 2010}}, {"name":1, "_id": 0, "offices.latitude":1, "offices.longitude":1, "total_money_raised": 1})
```

```python
list(companies_new_york)
```

```python
df = pd.DataFrame(collection.find(
    {"$and": [ # la condicion es que este en NY y haya sido fundada en 2005 o mas tarde
        {"offices.city": "New York"}, 
        {"founded_year": {"$gt": 2010}},
    ]},
    {"name":1,
     "category_code":1,
     "founded_year":1,
     "offices.city":1,
     "offices.latitude":1,
     "offices.longitude":1,
     "total_money_raised":1,
     "_id":0, }))
df.head(2)
```

Dentro de la columna de offices tengo acumulados "city", "latitude" y "longitude". 
Los separo en columnas distintas para unirlas despues.

```python
df.offices.head(2)
```

```python
city = []

for i in list(df.offices):
    city.append((i[0]['city']))
```

```python
df_city = pd.DataFrame(city)
df_city.columns = ["city"]
df_city.head(2)
```

```python
latitude = []

for i in list(df.offices):
    latitude.append((i[0]['latitude']))
```

```python
df_latitude = pd.DataFrame(latitude)
df_latitude.columns = ["latitude"]
df_latitude.head(2)
```

```python
longitude = []

for i in list(df.offices):
    longitude.append((i[0]['longitude']))
```

```python
df_longitude = pd.DataFrame(longitude)
df_longitude.columns = ["longitude"]
df_longitude.head(2)
```

Uniré las columnas que he creado junto al dataframe original para que todos los datos queden bien ordenados.

```python
df_union = pd.concat([df, df_city, df_latitude, df_longitude], axis=1)
```

```python
df_union.head(3)
```

```python
df_final = df_union.drop(df_union.columns[[1,2,3,4]], axis = 1)
```

```python
df_final.head(3)
```

El resultado es una tabla mas ordenada, pero aun queda por quitar los valores Nan y despejar la columna de total_money_raised hasta que sea unicamente numerica. 

```python
df_1 = df_final.dropna()
df_1.head(3)
```
