```python
import folium
from folium import Choropleth, Circle, Marker, Icon, Map
from folium.plugins import HeatMap, MarkerCluster
import pandas as pd
```

```python
df_companies = pd.read_csv('Data_companies.csv')
```

```python
df_companies1 = df_companies.rename(columns={'latitude':'latitud',
                                   'longitude':'longitud'})
```

```python
df_companies1
```

```python
df_starbucks=pd.read_csv('Data_starbucks.csv')
```

```python
df_starbucks.head(2)
```

```python
df_vegan=pd.read_csv('Data_vegan.csv')
```

```python
df_vegan.head(2)
```

```python
df_stadium = pd.read_csv('Data_stadium.csv')
```

```python
df_stadium.head(2)                             40.755665    -73.985542
```

```python
df_union = pd.concat([df_companies1, df_starbucks, df_vegan, df_stadium], axis=0)
```

```python
df_union
```

```python
latitud_inicial = 40.742665
longitud_inicial = -73.995542
```

```python
map_1 = Map(location=[latitud_inicial,longitud_inicial], zoom_start=15)
```

```python
sede = Marker(location=[latitud_inicial,longitud_inicial], tooltip = "Posible sede")
sede.add_to(map_1)
map_1
```

```python
for i,row in df_union.iterrows():
    #popup distrito
    etiqueta = {
        "location" : [row["latitud"], row["longitud"]],
        "tooltip" : row["etiqueta"]
    }
    
    if row["etiqueta"] == "Starbucks":
        icon = Icon( color = "pink",
                    prefix = "fa",
                    icon = "glass",
                    icon_color = "black"
        )
        
    elif row["etiqueta"] == "stadium":
        icon = Icon( color = "orange",
                    prefix = "fa",
                    icon = "glass",
                    icon_color = "black"
        )
        
    elif row["etiqueta"] == "companies":
        icon = Icon( color = "red",
                    prefix = "fa",
                    icon = "glass",
                    icon_color = "black"
        )
    
    else:
        icon = Icon( color = "green",
                    prefix = "fa",
                    icon = "briefcase",
                    icon_color = "black"
        )
    Marker (**etiqueta,icon = icon).add_to(map_1)
map_1
```
